## Wifi

```python
path = r"/Users/phurtado/Downloads/sitios-de-wifi.csv"
df = pd.read_csv(path)
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.info()
```

```python
pd.unique(df["comuna"])
```

```python
df_clean = df.dropna(subset=["comuna"])
print(pd.unique(df_clean["comuna"]))
print(df_clean.info())
```

```python
df_wifi = df_clean[["comuna","id"]].groupby("comuna").count()
```

# Merge

```python
df_merge = df_sedes.merge(df_centros,
                          how="left",
                          left_on="comuna",
                          right_on="comuna")
df_merge["id_usig"] = df_merge["id_usig"].fillna(0.0)
display(df_merge)
```

```python
df_merge_2 = df_merge.merge(df_wifi,
                            how="left",
                            on="comuna")
df_merge_2.columns = ["id","nombre","objeto", "domicilio", "comuna", "cantidad_centros", "cantidad_puntos_wifi"]
display(df_merge_2)
```

```python
df_merge_2.drop(["comuna"],axis=1,inplace=True)
display(df_merge_2)
```

```python
df_merge_2.to_csv("/Users/phurtado/Downloads/salida.csv",index=False)
```
